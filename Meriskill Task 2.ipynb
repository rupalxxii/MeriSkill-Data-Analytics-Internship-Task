# Outlier Removal

```python
avg = df["BloodPressure"].mean()
df["BloodPressure"] = df["BloodPressure"].apply(lambda v:avg if v==0 else v)
```

```python
plt.figure(figsize = (8,5))
sns.distplot(df["BloodPressure"] , bins = 30,color = "Orange")
plt.title(" Distribution of BloodPressure")
```

```python
bmi_avg = df["BMI"].mean()
df["BMI"] = df["BMI"].apply(lambda v:avg if v==0 else v)
```

```python
plt.figure(figsize = (8,5))
sns.displot(df, x="BMI",color = "Purple",kind= "kde")
plt.title(" Distribution of BMI")
```

As seen earlier there are amount whivh varies largely from rest of amount .It may not me outlier in real case bot for better modelling we need to remove it.
In above varible i replaced outlier value with mean. Noe lets use z-score technique to see outlier in other feature variable.

```python
df["z_score_bp"]=(df.BloodPressure-df.BloodPressure.mean()/df.BloodPressure.std())
df["z_score_Glucose"]=(df.Glucose-df.Glucose.mean()/df.Glucose.std())
```

```python
df.head()
```

```python
df.describe()
```

```python
df.drop(columns=["z_score_bp",  "z_score_Glucose"],inplace=True)
df.head()
```
